```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the file

raw_counts = pd.read_csv(r'C:\snowflake\raw_counts.csv')

print(raw_counts.columns.values)
raw_counts.describe()
```

```python
# Check for missing values
raw_counts.isnull().sum()
```

```python
# Subset the data bsed on what we want to analyze we will check what is the proportion of the cycles vs total trips
raw_counts_cycles = raw_counts[['count_point_id', 'year', 'hour', 'date', 'lat', 'long', 'bicycles', 'all_trips']]
```

```python
raw_counts_cycles['date'] = pd.to_datetime(raw_counts_cycles['date'])
```

```python
raw_counts_cycles.dtypes
```

```python
raw_counts_cycles['percentage_cyclists'] = raw_counts_cycles['bicycles'] / raw_counts_cycles['all_trips']

raw_counts_cycles
```

```python
# Subset the dataframe to find out which 10 countpoints have the highest and the lowest number of cyclists
raw_counts_cycles_head = raw_counts_cycles[['count_point_id', 'bicycles', 'all_trips']]

# Get the top 20 countpoints by total number of cyclists
grouped = raw_counts_cycles_head.groupby('count_point_id').sum()
```

```python
# Remove all the count points without bicycle counts from the dataframe
grouped[grouped['bicycles'] != 0].sort_values('bicycles', ascending = False).head(20)
```

```python
# Capture the countpoints with the highest bike volume
high_cyclecountpoints = [36109, 46107, 18486, 942947, 48553, 48552, 
                        56614, 942492, 6751, 28123, 8475, 56592,
                        36823, 56815, 26185, 46783, 27236, 18454,
                        8511, 47768]

# Subset the raw_counts_cyclist dataframe for the high volume countpoints
raw_counts_cycles_top = raw_counts_cycles[raw_counts_cycles['count_point_id'].isin(high_cyclecountpoints)]
raw_counts_cycles_top.head(20)
```

```python
# Retrieve the countpoints with low bike frequency but at least 1000 over the timeframe that is covered in the file
grouped[grouped['bicycles'] > 1000].sort_values('bicycles', ascending = True).head(20)
```

```python
# Capture the countpoints with the low bike volumes
low_cyclecountpoints = [942664, 942487, 36113, 47604, 16838,
                        18599, 26654, 930282, 57117, 36693,
                        942633, 27673, 56253, 56159, 38600,
                        48814, 36928, 942972]

raw_counts_cycles_bottom = raw_counts_cycles[raw_counts_cycles['count_point_id'].isin(low_cyclecountpoints)]
raw_counts_cycles_bottom.head(20)
```
